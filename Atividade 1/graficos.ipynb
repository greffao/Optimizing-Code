```python
import pandas as pd
import matplotlib.pyplot as plt

cores = ['#00008B',  # Dark Blue
         '#ADD8E6',  # Light Blue
         '#006400',  # Dark Green
         '#90EE90',  # Light Green
         '#8B0000',  # Dark Red
         '#FF3333']  # Light Red
```

```python
df = pd.read_csv('data.csv')
df
```

```python
# Medias task clock
media_task_clock = df.groupby('experimento')['task clock (msec)'].mean()
media_l1_dcache_load = df.groupby('experimento')['l1-dcache-load'].mean()
media_l1_dcache_load_misses = df.groupby('experimento')['l1-dcache-load-misses'].mean()
media_branch_instructions = df.groupby('experimento')['branch-instructions'].mean()
media_branch_misses = df.groupby('experimento')['branch-misses'].mean()

# Crie um DataFrame com todas as médias
dados = {
    'Experimento': media_task_clock.index,
    'Task Clock': media_task_clock.values,
    'L1 Dcache Load': media_l1_dcache_load.values,
    'L1 Dcache Load Misses': media_l1_dcache_load_misses.values,
    'Branch Instructions': media_branch_instructions.values,
    'Branch Misses': media_branch_misses.values
}

df_medias = pd.DataFrame(dados)

# Exiba o DataFrame com todas as médias
df_medias
```

```python
plt.bar(df_medias['Experimento'], df_medias['Task Clock'], 0.5, color=cores)

plt.xticks(rotation=90, fontsize=8)
plt.ylabel('Tempo Médio  (ms)')
plt.title('Tempo Médio em Milissegundos por Tipo de Multiplicação')

plt.show
```

```python
plt.bar(df_medias['Experimento'], df_medias['L1 Dcache Load'], 0.5, color=cores)

plt.xticks(rotation=90, fontsize=8)
plt.ylabel('Quantidade de Acessos a L1')
plt.title('Quantidade de Acessos a L1 para cada Tipo de Multiplicação')

plt.show
```

```python
plt.bar(df_medias['Experimento'], df_medias['L1 Dcache Load Misses'], 0.5, color=cores)

plt.xticks(rotation=90, fontsize=8)
plt.ylabel('Quantidade de Misses na L1')
plt.title('Quantidade de Misses na L1 para cada Tipo de Multiplicação')

plt.show
```

```python
plt.bar(df_medias['Experimento'], df_medias['Branch Instructions'], 0.5, color=cores)

plt.xticks(rotation=90, fontsize=8)
plt.ylabel('Quantidade de Branch')
plt.title('Quantidade de Instruções de Desvio (Branch)\n para cada Tipo de Multiplicação')

plt.show
```

```python
plt.bar(df_medias['Experimento'], df_medias['Branch Misses'], 0.5, label='Task Clock', color=cores)

plt.xticks(rotation=90, fontsize=8)
plt.ylabel('Quantidade de Branch Misses')
plt.title('Quantidade de Previsões de Desvios Erradas (Branch Misses)\n para cada Tipo de Multiplicação')

plt.show
```
